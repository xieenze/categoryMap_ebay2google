```python
from tqdm import trange
import numpy as np
import pandas as pd
from pprint import pprint
```

### 读取 ebay category数据,获得 us 数据

```python
ebay_category_path = './data/ares-presto_run_4_stmt_1_0.csv'
df = pd.read_csv(ebay_category_path,sep=',')
df = df[df["site_id"] == 0]
print(df.columns)
print('total nums of table:',len(df))
df.head()
```

```python
df_ebay_name=df['leaf_categ_name']
arr=np.array(df_ebay_name)
root_cate = [arr[i].split(':')[0].strip() for i in range(len(arr))]
distinct = np.unique(root_cate) 
distinct.shape
```

### 打印ebay的first category


```python
for i in distinct:
    print(i)
```

### 获取leaf_id2gpcid表的 us 数据


```python
df2 = pd.read_csv("./data/leaf_id2gpcid.tsv",sep='\t')
df2 = df2.query('site_id==0')
print(len(df2))
df2.head(10)
```

### select df.*,df2.gpc_id from df left join df2 on df.leaf_categ_id=df2.leaf_categ_id  
### 过滤掉gpc_id为空的

```python
df_new = pd.merge(df,df2[["gpc_id"]],how='right',left_on=df.leaf_categ_id,right_on=df2.leaf_categ_id)
filtered_df = df_new[df_new['leaf_categ_id'].notnull()]
```

```python
filtered_df.head(10)
```

### 读取gpc_id2name 的 us 数据


```python
df3 = pd.read_csv('./data/gpc_id2name.tsv',sep='\t')
```

### df_finn为处理后的表，包含ebay_cate 和 gg_cate一一对应关系

```python
df_fin = pd.merge(filtered_df,df3[["GPC_NAME"]],how='left',left_on=filtered_df.gpc_id,right_on=df3.GPC_ID)
df_finn = df_fin[['leaf_categ_name','GPC_NAME']]
df_finn =df_finn[df_finn['GPC_NAME'].notnull()]
df_finn.head(10)
```

```python
tmp = df_finn[100:110].values
for i in tmp:
    print(i)
```

```python
arr = df_finn.values
for i in range(len(arr)):
    arr[i][1] =  arr[i][1].split(">")[0].strip()
    #print(arr[i][1].split(">")[0])
```

### df_100为最终表。包含ebay_cate和gg_first_cate对应关系

```python
df_100 = pd.DataFrame(arr,columns=['ebay_categ',"gg_first_categ"])
```

```python
df_100=df_100.sort_values(by=['gg_first_categ'])
df_100
```

```python
np.array(df_100.query("gg_first_categ=='Toys & Games'")[['ebay_categ']][0:10])
```

```python
gg_fitst_cate = np.array(pd.read_table('./data/gg_first_category.txt',header=None))
gg_fitst_cate = np.reshape(gg_fitst_cate,21)
gg_fitst_cate
for i in gg_fitst_cate:
    print (i)
```

### 查询每个google root category对应哪些ebay category，并分别保进txt中去

```python
query_sql = "gg_first_categ=='need_repalece'"
for i in gg_fitst_cate:
    real_query_sql = query_sql.replace("need_repalece",i)
    #print(query_sql1)
    df_each_gg_first_cate = df_100.query(real_query_sql)[['ebay_categ']]
    df_each_gg_first_cate.to_csv("./data/dataset/"+i+'.txt',header=False,index=False)
    print(i,":",len(df_each_gg_first_cate))
```
