```python
import gensim
from nltk.tokenize import word_tokenize
from nltk.tokenize import wordpunct_tokenize
from tqdm import trange
import numpy as np
import pandas as pd

```

```python
#读取 google category数据
df = pd.read_csv('./data/gpc_id2name.tsv',sep='\t')
```

```python
#dataframe2numpy  第2列
arr = df.values[:,1:2]
new_arr = arr.reshape(len(arr))
new_list = list(new_arr)
```

```python
new_new_list = [i.replace(">"," ").replace("&"," ") for i in new_list]
```

### 1. 引入文本处理库gensim，读取 google category数据。

```python
print("Number of documents:",len(new_new_list))
```

### 2. 使用NLTK进行分词。

```python
gen_docs = [[w.lower() for w in word_tokenize(text)] 
            for text in new_new_list]


for i in range(20):
    print(gen_docs[i])
```

### 3. 创建词典，映射所有单词。

```python
dictionary = gensim.corpora.Dictionary(gen_docs)
```

```python
for i in range(len(dictionary)):
    print(i, dictionary[i])
```

### 4. 创建词袋模型语料库，记录每个文本的词频。

```python
corpus = [dictionary.doc2bow(gen_doc) for gen_doc in gen_docs]
for corp in corpus:
    print(corp)
```

### 5. 从语料库中生成tf-idf模型。最后打印出的`num_nnz`是词的总量。

```python
tf_idf = gensim.models.TfidfModel(corpus)
```

```python
print(tf_idf)
tf_idf.save('./gg_category.tfidf_model')
```
